```python
def load_data():

    # Replacing missing values with median of that specific column
    df['Year'].fillna(value=df['Year'].median(), inplace=True)

    x = df.drop('Rape', axis=1).values
    y = df['Rape'].values

    return x, y, df
```

```python
import numpy as np
import pandas as pd

def normalize(data):
    # Convert data to a pandas DataFrame for easy column-wise processing
    df = pd.DataFrame(data)

    # Identify non-numeric columns
    non_numeric_columns = df.select_dtypes(exclude=np.number).columns

    # One-hot encode non-numeric columns
    df_encoded = pd.get_dummies(df, columns=non_numeric_columns)

    # Convert the DataFrame back to a NumPy array
    data_encoded = df_encoded.to_numpy()

    min_val = np.min(data_encoded, axis=0)
    max_val = np.max(data_encoded, axis=0)

    # Avoid division by zero
    range_val = max_val - min_val
    range_val[range_val == 0] = 1

    norm_x = (data_encoded - min_val) / range_val
    return norm_x
```

```python
import numpy as np
import matplotlib.pyplot as plt

def normalize_features(features):
    mean = np.mean(features, axis=0)
    std = np.std(features, axis=0)
    normalized_features = (features - mean) / std
    return normalized_features, mean, std

def linear_regression(features, labels, learning_rate, epochs):
    m, n = features.shape
    thetas = np.zeros(n)  # Initialize parameters to zeros
    cost_history = []

    for epoch in range(epochs):
        # Hypothesis
        predictions = np.dot(features, thetas)

        # Calculate error
        errors = predictions - labels

        # Update parameters
        thetas -= (learning_rate / m) * np.dot(errors, features)

        # Calculate cost (mean squared error)
        cost = np.sum(errors ** 2) / (2 * m)
        cost_history.append(cost)

    return thetas, cost_history

# Assuming x_train, y_train are your training features and labels
x_train = np.array([[2001], [2002], [2003], [2004]])  # Example data, replace with your own
y_train = np.array([100, 120, 140, 160])  # Example labels, replace with your own

# Normalize features
x_train_normalized, mean, std = normalize_features(x_train)

alpha = 0.01
epochs = 1000

# Add a column of ones to the features for the bias term
x_train_bias = np.c_[np.ones(x_train_normalized.shape[0]), x_train_normalized]

# Apply linear regression
thetas, cost_history = linear_regression(features=x_train_bias, labels=y_train, learning_rate=alpha, epochs=epochs)

# Visualize the cost over epochs
plt.plot(range(1, epochs + 1), cost_history, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Cost')
plt.title('Cost Function Over Epochs')
plt.show()

# Print the final parameters
print("Final Parameters (Thetas):", thetas)
```

```python
import numpy as np

def linear_regression_cost(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    squared_errors = np.square(predictions - y)
    cost = 1/(2*m) * np.sum(squared_errors)
    return cost

# Assuming X is the feature matrix (with a column of ones for the intercept term)
# and y is the target variable (Total_cases)

# Add a column of ones to X for the intercept term
X = np.column_stack((np.ones(len(X)), X))

# Example parameter vector [theta0, theta1]
theta = np.zeros(X.shape[1])

# Trim y to match the number of examples in X
y = y[:X.shape[0]]

# Calculate the cost
cost = linear_regression_cost(X, y, theta)

print(f"Cost: {cost}")
```
